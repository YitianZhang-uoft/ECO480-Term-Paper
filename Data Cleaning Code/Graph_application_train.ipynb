```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# application_train - First and Main dataset
application_train = pd.read_csv('application_train.csv')
```

```python
application_train
```

```python
missing_values_count = application_train.isnull().sum().sum()
missing_values_count
```

```python
application_train.describe()
```

```python
variable_types = application_train.dtypes.value_counts()
variable_types
```

```python
type(application_train['NAME_CONTRACT_TYPE'][1])
```

```python
# Identification if loan is cash or revolving vs Target

grouped_data = application_train.groupby('NAME_CONTRACT_TYPE')['TARGET'].value_counts().unstack()

# Plotting the graph
grouped_data.plot(kind='bar', stacked=True)

# Labeling the axes and providing a title
plt.xlabel('NAME_CONTRACT_TYPE')
plt.ylabel('Count')
plt.title('Distribution of TARGET by NAME_CONTRACT_TYPE')

# Adding a legend
plt.legend(title='TARGET')

# Displaying the graph
plt.show()
```

```python
# Gender vs Target

grouped_data = application_train.groupby('CODE_GENDER')['TARGET'].value_counts().unstack()

# Plotting the graph
grouped_data.plot(kind='bar', stacked=False)

# Labeling the axes and providing a title
plt.xlabel('CODE_GENDER')
plt.ylabel('Count')
plt.title('Distribution of TARGET by GENDER')

# Adding a legend
plt.legend(title='TARGET')

# Displaying the graph
plt.show()
```

```python
# Income vs Target

income_categories = pd.cut(application_train['AMT_INCOME_TOTAL'], bins=8)
# Assign the income categories as a new column in the DataFrame
application_train['AMT_INCOME_TOTAL'] = income_categories

category_counts = application_train.groupby(['AMT_INCOME_TOTAL', 'TARGET']).size().unstack(fill_value=0)
category_counts
```
