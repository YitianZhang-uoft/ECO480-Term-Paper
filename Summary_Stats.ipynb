```python
import pandas as pd
```

```python
# application_train - First Dataset
application_train = pd.read_csv('application_train.csv')
```

```python
# Missing Data
missing = pd.DataFrame(application_train.isnull().sum())
missing
```

```python
# Summary Statistics
X_describe = application_train.describe(include='all').T

result = pd.concat([X_describe], axis=0).rename(columns={'count': 'Non-missing Count', 'mean': 'Mean', 'std': 'Standard Deviation', 'min': 'Minimum', '25%': '25th percentile', '50%': '50th percentile', '75%': '75th percentile', 'max': 'Maximum', 'unique': 'Unique Values Count', 'top': 'Mode', 'freq': 'Mode Frequency'})
result['Missing Count'] = pd.concat([missing], axis=0)
result['Data Type'] = pd.concat([application_train.dtypes], axis=0)

# Display the result
pd.set_option('display.max_rows', None)
display(result)
```
